```python
url = 'http://nesssi.cacr.caltech.edu/cgi-bin/getcssconedbid_release2.cgi'
values = {'Name' : 'Her X-1',
          'Radius': '.1',
         '.submit' : 'Submit',
         'DB' : 'photcat',
         'OUT' : 'web',
         'SHORT' : 'short',
         'PLOT' : 'plot'}
data = urllib.parse.urlencode(values) 
data = data.encode('ascii')
req = urllib.request.Request(url, data)
resp = urllib.request.urlopen(req)
output = resp.read()
print(output)
print(type('output'))
```

```python
stringoutput = output.decode('ASCII')
print(stringoutput)
```

```python
spl = stringoutput.split('data:')
data0= spl[1]
print(data0)
```

```python
data1 = data0.split('}')[0]
print(data1)
```

```python
print(type(data1))
```

```python
def extr(lst,el): 
    return [item[el] for item in lst]
x = extr(data1,0)
y = extr(data1,1)
yerror = extr(data1,2)
import matplotlib.pyplot as plt
plt.errorbar(x,y,yerr=yerror,fmt='o')
plt.title('Her X-1 Light Curve')
plt.xlabel('Date (MJD)')
plt.ylabel('V mag')
```

```python
url = 'http://nesssi.cacr.caltech.edu/cgi-bin/getcssconedbid_release2.cgi'
values = {'Name' : 'Her X-1',
          'Radius': '.1',
         '.submit' : 'Submit',
         'DB' : 'photcat',
         'OUT' : 'vot',
         'SHORT' : 'short',
         'PLOT' : 'plot'}
data = urllib.parse.urlencode(values) 
data = data.encode('ascii')
req = urllib.request.Request(url, data)
resp = urllib.request.urlopen(req)
votoutput = resp.read()
print(votoutput)
print(type('votoutput'))
```

```python
vot1 = votoutput.decode().split('type="hidden" name="url" value="')[1]
print(vot1)
```

```python
vot2 = vot1.split('"')[0]
print(vot2)
```

```python
import shutil
import tempfile
import urllib.request
import astropy.io.votable
from astropy.io.votable import parse

with urllib.request.urlopen(vot2) as response:
    with tempfile.NamedTemporaryFile(delete=False) as tmp_file:
        shutil.copyfileobj(response, tmp_file)
        votable=parse(tmp_file.name, pedantic = False)
```

```python
table = votable.get_first_table()
print(table)
```

```python
datav = table.array
print(datav)
print(datav.shape)
```
